# Enriquecimiento de la dataset

```python
compras_df = pd.read_csv("/content/drive/MyDrive/Maestria CIENCIA DE DATOS/Inteligencia Artificial/Unidad 2/Tarea grupo/notebooks/data/clean_compras_2016.csv")
compras_df.head(3)
```

```python
compras_df.dtypes
```

##a.  Expansión de la fecha

```python
compras_df["FECHA"] = pd.to_datetime(compras_df["FECHA"])
```

```python
compras_df["DIA"] = compras_df["FECHA"].apply(lambda x: x.day)
```

```python
compras_df["MES"] = compras_df["FECHA"].apply(lambda x: x.month)
```

```python
compras_df["ANIO"] = compras_df["FECHA"].apply(lambda x: x.year)
```

```python
compras_df.head(3)
```

## b. Proyección de una variable diferente a la vista en clase

## Número de compras por mes

```python
compras_por_mes = compras_df["MES"].value_counts().to_dict()
compras_por_mes
```

```python
list(compras_por_mes.keys())
```

```python
compras_df['COMPRAS_POR_MES'] = compras_df.apply(lambda row: compras_por_mes[(row['MES'])], axis=1)
compras_df.head(3)
```

```python
compras_df.tail(3)
```

## Importe promedio por dia del mes

```python
dia_mean_compras = compras_df.groupby("DIA")["IMPORTE"].mean()
dia_mean_compras
```

```python
dia_mean_compras.to_dict()
```

```python
compras_df['IMPORTE_PROMEDIO_POR_DIA'] = compras_df.apply(lambda row: dia_mean_compras[(row['DIA'])], axis=1)
compras_df.head(3)
```
