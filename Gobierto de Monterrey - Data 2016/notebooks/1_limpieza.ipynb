# De febrero a diciembre
Los registros de febrero a diciembre presentar más similitudes que los registros de enero, por lo que se ha realizado la limpieza de todos estos meses.

```python
########################################
# ANALIZANDO LAS COLUMNAS POR SHEETS
########################################

# CONTRATISTAS: A continuación se presentan los nombres de las columnas correspondientes a Contratistas
contratistas_names = ['PROVEEDORES', 'FDOS FED Y CONTRATISTAS', ' FDO FED Y CONTRATI', 'CONTRAT Y FDOS FED', 'FDO FED Y CONTRAT', 'FDOS FED Y CONTRAT', 'CONTRATISTAS Y F FED', 'CONTRATISTAS Y FD FED',  'OP Y FDOS FED', 'FDOS FED Y CONTRATISTAS', 'FDS FED Y CONTRAT',  'CONTRAT Y FDOS FED']
# GASTOS VARIOS: A continuación se presentan los nombres de las columnas correspondientes a Gastos Varios
gastos_names = ["GASTOS DE VIAJE Y REPRESENTACION", "GTS VARIOS", 'GTS GRALES']
# SERVICIOS PROFESIONALES: A continuación se presentan los nombres de las columnas correspondientes a Servicios Profesionales
servicios_prof_names = ['SERV PROF', 'SERVPROF']
# COMUNICACION: : A continuación se presentan los nombres de las columnas correspondientes a Comunicación
comunicacion_names = ['COMUNICACION', 'COMUNIC']
# GASTOS REPRESENTACION: A continuación se presentan los nombres de las columnas correspondientes a Gastos de Representación
gastos_repre_names = ['GTS REPRE']
# SERVICIOS PERSONALES: A continuación se presentan los nombres de las columnas correspondientes a Servicios Personales
servicios_personales_names = ['SERV PERS']
```

```python
# Lista para almacenar todos los DataFrames
lista_dfs = []

# Listar todos los archivos en la carpeta
archivos_excel = [f for f in os.listdir(ruta_carpeta) if f.endswith('.xlsx')]

# Iterar sobre los archivos Excel y obtener sus hojas
for archivo in archivos_excel:
    if archivo == "PAGOS DE ENERO DEL 2016.xlsx":  # Excluyendo este archivo
        continue
    ruta_archivo = os.path.join(ruta_carpeta, archivo)

    try:
        # Leer las hojas del archivo Excel
        excel_file = pd.ExcelFile(ruta_archivo)
        sheets = excel_file.sheet_names
        #print(f"Archivo: {archivo} tiene {len(sheets)} hojas: {sheets}")

        # Iterar sobre cada hoja y leer los datos saltando las primeras 5 filas
        new_col_name = None
        for sheet in sheets:
            print(f"Datos de la hoja '{sheet}' del archivo '{archivo}':")
            df = pd.read_excel(ruta_archivo, sheet_name=sheet, skiprows=5)

            # Convertir los nombres de las columnas a mayúsculas
            df.columns = df.columns.str.upper()

            # Renombrar columnas específicas según tu mapeo
            df = df.rename(columns={
                'FECHAEMIS': 'FECHA',
                'FACTURA': 'FACTURA',
                'NOMBRE': 'NOMBRE',
                'RFC': 'RFC',
                'DESCRIPCION': 'CONCEPTO',
                'VALOR': 'IMPORTE'
            })

            # Asignar el nombre correcto basado en el sheet
            if sheet in contratistas_names:
                new_col_name = "CONTRATISTAS"
            elif sheet in gastos_names:
                new_col_name = "GASTOS VARIOS"
            elif sheet in servicios_prof_names:
                new_col_name = "SERVICIOS PROFESIONALES"
            elif sheet in comunicacion_names:
                new_col_name = "COMUNICACION"
            elif sheet in gastos_repre_names:
                new_col_name = "GASTOS REPRESENTACION"
            elif sheet in servicios_personales_names:
                new_col_name = "SERVICIOS PERSONALES"

            if new_col_name is not None:
                # Agregar la nueva columna con el valor de new_col_name
                df["SHEET"] = new_col_name

            # Añadir el DataFrame a la lista
            lista_dfs.append(df)

    except Exception as e:
        print(f"Error al leer el archivo {archivo}: {e}")

# Concatenar todos los DataFrames en uno solo
df_concatenado = pd.concat(lista_dfs, ignore_index=True)
df_feb_to_dic = df_concatenado.copy()[['FECHA','FACTURA','NOMBRE','RFC','CONCEPTO','IMPORTE',"SHEET"]]
print(f"Elementos antes de eliminar nulos: {len(df_feb_to_dic)}")
df_feb_to_dic = df_feb_to_dic.dropna(how='any')
print(f"Elementos después de eliminar nulos: {len(df_feb_to_dic)}")
```
